```python
# ═══════════════════════════════════════════════════════════════════════
# 0. CONFIGURATION & COLORS
# ═══════════════════════════════════════════════════════════════════════
COLOR_REV         = '#2E86AB'
COLOR_PROFIT      = '#6BBF59'
COLOR_LOSS_ORG    = '#A8312C'
COLOR_LOSS_PROMO  = '#ED8936'
plt.rcParams['font.family'] = 'sans-serif'

# ═══════════════════════════════════════════════════════════════════════
# 1. DATA LOADING & MERGING (As per Datathon v2 (1).pdf Data Map)
# ═══════════════════════════════════════════════════════════════════════
orders = pd.read_csv('/content/gdrive/MyDrive/Datathon/data/orders.csv', parse_dates=['order_date'])
items = pd.read_csv('/content/gdrive/MyDrive/Datathon/data/order_items.csv', low_memory=False)
products = pd.read_csv('/content/gdrive/MyDrive/Datathon/data/products.csv')

df = items.merge(products[['product_id', 'cogs']], on='product_id', how='left')
df = df.merge(orders[['order_id', 'order_date']], on='order_id', how='left')

# Calculating core metrics
df['net_rev'] = df['quantity'] * df['unit_price']
df['total_cogs'] = df['quantity'] * df['cogs']
df['gross_profit'] = df['net_rev'] - df['total_cogs']
df['is_promo'] = df['promo_id'].notna().astype(int)

# ═══════════════════════════════════════════════════════════════════════
# 2. KPI CALCULATIONS (FILTERED 2013 - 2022)
# ═══════════════════════════════════════════════════════════════════════
df_filtered = df[(df['order_date'].dt.year >= 2013) & (df['order_date'].dt.year <= 2022)].copy()

# Monthly Aggregation
monthly = df.set_index('order_date').resample('ME').agg({
    'net_rev': 'sum',
    'gross_profit': 'sum',
    'is_promo': 'mean'
}).reset_index()

monthly['ym_dt'] = monthly['order_date']
monthly['margin_pct'] = (monthly['gross_profit'] / monthly['net_rev'] * 100).round(2)

# CAGR Calculation (n = 9 years)
rev_2013 = monthly[monthly['ym_dt'].dt.year == 2013]['net_rev'].sum()
rev_2022 = monthly[monthly['ym_dt'].dt.year == 2022]['net_rev'].sum()
cagr = (np.power(rev_2022 / rev_2013, 1/9) - 1) * 100

# Peak Quarter Calculation
df_filtered['quarter'] = df_filtered['order_date'].dt.quarter
q_rev = df_filtered.groupby('quarter')['net_rev'].sum()
peak_q = q_rev.idxmax()
peak_q_pct = (q_rev.max() / q_rev.sum()) * 100

pos_months = (monthly['gross_profit'] > 0).sum()
neg_months = (monthly['gross_profit'] <= 0).sum()

# ═══════════════════════════════════════════════════════════════════════
# 3. VISUALIZATION
# ═══════════════════════════════════════════════════════════════════════
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(15, 10), sharex=True, gridspec_kw={'height_ratios': [3, 2]})
plt.subplots_adjust(hspace=0.12)

# --- Subplot 1: Trends ---
ax1.fill_between(monthly['ym_dt'], monthly['net_rev'], alpha=0.15, color=COLOR_REV, label='Net Revenue')
ax1.plot(monthly['ym_dt'], monthly['net_rev'], color=COLOR_REV, linewidth=2, alpha=0.7)

ax1.fill_between(monthly['ym_dt'], monthly['gross_profit'], 0,
                 where=(monthly['gross_profit'] >= 0), color=COLOR_PROFIT, alpha=0.5, label='Gross Profit (Gain)')
ax1.fill_between(monthly['ym_dt'], monthly['gross_profit'], 0,
                 where=(monthly['gross_profit'] < 0), color=COLOR_LOSS_ORG, alpha=0.6, label='Gross Profit (Loss)')

ax1.set_title('Financial Performance Analysis (2012-2022)', fontsize=15, fontweight='bold', loc='left', pad=25)
ax1.set_ylabel('Value (VND)', fontsize=11)
ax1.yaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f'{x/1e6:.0f}M'))
ax1.grid(True, linestyle='--', alpha=0.3)
ax1.legend(loc='lower left', frameon=True, fontsize=10, ncol=3)

# Annotation Box - Reordered: CAGR, Peak Q, Profitable, Loss
info_text = (f"Revenue CAGR (13-22): {cagr:.2f}%      \n"
             f"Peak Quarter (13-22): Q{peak_q} ({peak_q_pct:.1f}%)      \n"
             f"Profitable Months: {pos_months} months      \n"
             f"Loss Months: {neg_months} months      ")

props = dict(boxstyle='round,pad=1.0', facecolor='white', alpha=0.95, edgecolor='#CCCCCC')
ax1.text(0.98, 0.95, info_text, transform=ax1.transAxes, fontsize=11, fontweight='bold',
        verticalalignment='top', horizontalalignment='right',
        multialignment='left', bbox=props, color='#333333', linespacing=1.6)

# --- Subplot 2: Margin & Promo Impact ---
bar_colors = []
for _, row in monthly.iterrows():
    if row['margin_pct'] >= 0:
        bar_colors.append(COLOR_PROFIT)
    elif row['is_promo'] > 0.384: # Threshold for high promo penetration
        bar_colors.append(COLOR_LOSS_PROMO)
    else:
        bar_colors.append(COLOR_LOSS_ORG)

ax2.bar(monthly['ym_dt'], monthly['margin_pct'], color=bar_colors, width=22, alpha=0.8)
ax2.axhline(0, color='black', linewidth=1, alpha=0.5)
ax2.set_ylabel('Gross Margin (%)', fontsize=11)
ax2.set_ylim(monthly['margin_pct'].min() - 10, monthly['margin_pct'].max() + 10)
ax2.grid(True, axis='y', linestyle='--', alpha=0.3)

custom_lines = [Line2D([0], [0], color=COLOR_PROFIT, lw=4),
                Line2D([0], [0], color=COLOR_LOSS_ORG, lw=4),
                Line2D([0], [0], color=COLOR_LOSS_PROMO, lw=4)]
ax2.legend(custom_lines, ['Gain', 'Organic Loss', 'Promo-Driven Loss'], loc='lower left', ncol=3, fontsize=9)

plt.xlabel('Timeline (Monthly)', fontsize=11)
plt.tight_layout()
plt.show()

# ═══════════════════════════════════════════════════════════════════════
# 4. STRATEGIC INSIGHTS
# ═══════════════════════════════════════════════════════════════════════
display(Markdown(
    f"### 📌 **Strategic Business Analysis Summary**\n"
    f"* **Growth Metric**: The 9-year Revenue CAGR (2013-2022) is **{cagr:.2f}%**.\n"
    f"* **Seasonal Performance**: **Quarter {peak_q}** is identified as the peak period, contributing **{peak_q_pct:.1f}%** of total sales over the decade.\n"
    f"* **Profitability Diagnostics**: The analysis reveals **{neg_months} months** of net loss. Orange bars highlight periods where high promotion penetration (>38.4%) impacted margins.\n"
    f"* **Action Plan**: Management should strictly monitor discount levels during the high-traffic **Q{peak_q}** to ensure that top-line growth is not achieved at the expense of net profitability[cite: 1]."
))
```

```python
# ═══════════════════════════════════════════════════════════════════════
# CHART 2 — PROMO vs NON-PROMO (English Version with Grid & Clean Axes)
# ═══════════════════════════════════════════════════════════════════════
# Data aggregation for diagnostic analysis
overview = df.groupby('has_promo').agg(
    total_gross_rev = ('gross_rev', 'sum'),
    total_discount  = ('discount_amount', 'sum'),
    net_revenue     = ('net_rev', 'sum'),
    total_cogs      = ('total_cogs', 'sum'),
    gross_profit    = ('gross_profit', 'sum'),
).reset_index()

# Metric calculation
overview['margin_%']        = (overview['gross_profit'] / overview['net_revenue'] * 100).round(2)
overview['discount_rate_%'] = (overview['total_discount'] / overview['total_gross_rev'] * 100).round(2)
overview['label'] = overview['has_promo'].map({True: 'With Promo', False: 'No Promo'})

fig, axes = plt.subplots(1, 3, figsize=(18, 6))

# Define sub-plots configuration
for ax in axes:
    # Enable background grid
    ax.grid(True, linestyle='--', alpha=0.5, axis='y', zorder=0)
    # Only show left (Y) and bottom (X) spines for a clean look
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines['left'].set_linewidth(1.2)
    ax.spines['bottom'].set_linewidth(1.2)

# Panel 1: Gross Margin %
colors = [COLOR_LOSS if v < 0 else COLOR_PROFIT for v in overview['margin_%']]
axes[0].bar(overview['label'], overview['margin_%'], color=colors, width=0.4, zorder=3)
axes[0].set_title('Gross Margin %\n(Profit / Net Revenue)', fontweight='bold')
axes[0].set_ylabel('%')
axes[0].set_ylim(0, overview['margin_%'].max() * 1.25) # Padding for labels
for j, v in enumerate(overview['margin_%']):
    axes[0].text(j, v + 0.5, f'{v:.2f}%', ha='center', fontweight='bold')

# Panel 2: Net Revenue Comparison
axes[1].bar(overview['label'], overview['net_revenue']/1e6, color=[COLOR_PROMO, COLOR_REV], width=0.4, zorder=3)
axes[1].set_title('Net Revenue (Million VND)\n(Revenue AFTER Discount)', fontweight='bold')
axes[1].set_ylabel('Million VND')
axes[1].set_ylim(0, (overview['net_revenue'].max()/1e6) * 1.25)
for j, v in enumerate(overview['net_revenue']):
    pct = v / overview['net_revenue'].sum() * 100
    axes[1].text(j, v/1e6 + 50, f'{v/1e6:,.0f}M\n({pct:.1f}%)', ha='center', fontweight='bold', fontsize=9)

# Panel 3: Gross Profit
colors2 = [COLOR_LOSS if v < 0 else COLOR_PROFIT for v in overview['gross_profit']]
axes[2].bar(overview['label'], overview['gross_profit']/1e6, color=colors2, width=0.4, zorder=3)
axes[2].set_title('Gross Profit (Million VND)', fontweight='bold')
axes[2].set_ylabel('Million VND')
axes[2].set_ylim(0, (overview['gross_profit'].max()/1e6) * 1.25)
for j, v in enumerate(overview['gross_profit']):
    axes[2].text(j, v/1e6 + 50, f'{v/1e6:,.0f}M', ha='center', fontweight='bold')

plt.suptitle('Promo vs Non-Promo Performance Analysis', fontsize=14, fontweight='bold')
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()

# ── Auto Insights (English) ──
margin_promo    = overview[overview['has_promo']==True]['margin_%'].values[0]
margin_nonpromo = overview[overview['has_promo']==False]['margin_%'].values[0]

display(Markdown(
    f'**📌 Strategic Business Insights — Chart 2:**\n'
    f'- **Margin Impact**: Promotions drive volume but significantly erode profitability, reducing margin from **{margin_nonpromo:.2f}%** to **{margin_promo:.2f}%**.\n'
    f'- **Revenue Contribution**: "With Promo" orders account for approximately **{ (overview[overview["has_promo"]==True]["net_revenue"].values[0] / overview["net_revenue"].sum() * 100):.1f}%** of total net revenue.\n'
    f'- **Diagnostic**: High dependence on discounts (38.4% of orders) indicates a "price-sensitive" customer base that may risk long-term brand equity[cite: 1, 3].\n'
    f'- **Prescriptive Recommendation**: Implement a minimum floor price linked to COGS to prevent monthly gross margins from dipping into negative territory during flash sales.'
))
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.dates as mdates
import matplotlib.ticker as mticker

# --- BƯỚC 1: CHUẨN BỊ DỮ LIỆU ---
channel_colors = {
    'direct': '#1F3A5F',         # Navy
    'email_campaign': '#2E86AB',  # Blue
    'organic_search': '#5D6D7E',  # Gray-Blue
    'paid_search': '#6BBF59',     # Green
    'referral': '#AAB7B8',        # Silver
    'social_media': '#F1C40F'     # Yellow
}

# 1. Xử lý dữ liệu Web Traffic
fact_web['traffic_date'] = pd.to_datetime(fact_web['traffic_date'])
monthly_source = fact_web.groupby([pd.Grouper(key='traffic_date', freq='MS'), 'traffic_source'])['sessions'].sum().reset_index()
monthly_bounce = fact_web.groupby(pd.Grouper(key='traffic_date', freq='MS'))['bounce_rate'].mean().reset_index()

# Làm mượt Bounce Rate (Moving Average 3 tháng)
monthly_bounce['bounce_rate_smoothed'] = monthly_bounce['bounce_rate'].rolling(window=3, min_periods=1).mean()

# 2. Xử lý dữ liệu Khuyến mãi (Logic Quý chiếm đa số ngày)
def get_majority_quarter(row):
    all_days = pd.date_range(start=row['start_date'], end=row['end_date'])
    quarters = all_days.to_series().dt.quarter
    return quarters.mode()[0]

# Áp dụng logic cho bảng promotions
promotions['majority_quarter'] = promotions.apply(get_majority_quarter, axis=1)

# --- BƯỚC 2: TRỰC QUAN HÓA ---

# Khởi tạo Figure với 2 hàng (Subplots)
fig, (ax1, ax3) = plt.subplots(2, 1, figsize=(20, 14), dpi=100)
plt.subplots_adjust(hspace=0.4, bottom=0.15) # Khoảng cách giữa các biểu đồ

# ═══════════════════════════════════════════════════════════════════════
# BIỂU ĐỒ 1 (TRÊN): WEB TRAFFIC (Học theo style của bạn)
# ═══════════════════════════════════════════════════════════════════════
# Trục trái: Sessions
sns.lineplot(data=monthly_source, x='traffic_date', y='sessions', hue='traffic_source',
             ax=ax1, palette=channel_colors, linewidth=2.5, alpha=0.9)

ax1.set_ylabel('Sessions', fontsize=12, fontweight='bold')
ax1.set_xlabel('')
ax1.grid(axis='y', linestyle='-', alpha=0.2)

# Trục phải: Bounce Rate (Đã làm mượt)[cite: 1]
ax2 = ax1.twinx()
sns.lineplot(data=monthly_bounce, x='traffic_date', y='bounce_rate_smoothed', ax=ax2,
             color='red', linestyle='--', linewidth=2, label='Avg. Bounce Rate (Smoothed)', alpha=0.8)
ax2.set_ylabel('Avg. Bounce Rate', fontsize=12, color='red', fontweight='bold')
ax2.set_ylim(0, monthly_bounce['bounce_rate'].max() * 1.5)

# Định dạng trục X cho biểu đồ 1
ax1.xaxis.set_major_locator(mdates.MonthLocator(interval=3))
ax1.xaxis.set_major_formatter(mdates.DateFormatter('%b %Y'))
plt.setp(ax1.get_xticklabels(), rotation=45, ha='right')

# Legend cho biểu đồ trên (Nằm dưới biểu đồ 1)
lines1, labels1 = ax1.get_legend_handles_labels()
ax1.get_legend().remove()
ax1.legend(lines1, labels1, loc='upper center', bbox_to_anchor=(0.5, -0.22),
           ncol=6, frameon=False, fontsize=10)

ax1.set_title('Web Traffic · Sessions + Smoothed Bounce Rate', fontsize=16, loc='left', pad=20, fontweight='bold')
sns.despine(ax=ax1, top=True, right=False)

# ═══════════════════════════════════════════════════════════════════════
# BIỂU ĐỒ 2 (DƯỚI): PROMOTION DISTRIBUTION (Histogram)
# ═══════════════════════════════════════════════════════════════════════
sns.countplot(data=promotions, x='majority_quarter', ax=ax3, palette='magma', edgecolor='black')

ax3.set_title('Promotions Distribution',
              fontsize=16, loc='left', pad=20, fontweight='bold')
ax3.set_xlabel('Quarter', fontsize=12, fontweight='bold')
ax3.set_ylabel('Number of Promotions', fontsize=12, fontweight='bold')
ax3.set_xticklabels(['Q1', 'Q2', 'Q3', 'Q4'])

# Hiển thị số lượng chính xác trên đầu cột
for p in ax3.patches:
    ax3.annotate(f'{int(p.get_height())}', (p.get_x() + p.get_width() / 2., p.get_height()),
                 ha='center', va='center', fontsize=12, color='black', xytext=(0, 8),
                 textcoords='offset points', fontweight='bold')

sns.despine(ax=ax3)
plt.show()
```
